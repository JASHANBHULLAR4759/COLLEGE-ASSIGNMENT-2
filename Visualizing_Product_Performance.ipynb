```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean data
sales_df = pd.read_csv('sales_data.csv')
sales_df['Unit_Price'] = sales_df['Unit_Price'].replace('[\$,]', '', regex=True).astype(float)
sales_df['Total_Sale'] = sales_df['Total_Sale'].replace('[\$,]', '', regex=True).astype(float)

# Set style
sns.set(style="whitegrid")

# 1. Top 5 by Revenue
top_revenue = sales_df.groupby('Product')['Total_Sale'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_revenue.values, y=top_revenue.index, palette="Blues_d")
plt.title("💰 Top 5 Products by Total Revenue")
plt.xlabel("Total Revenue ($)")
plt.tight_layout()
plt.show()

# 2. Top 5 by Quantity
top_quantity = sales_df.groupby('Product')['Quantity'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_quantity.values, y=top_quantity.index, palette="Greens_d")
plt.title("📦 Top 5 Products by Quantity Sold")
plt.xlabel("Units Sold")
plt.tight_layout()
plt.show()

# 3. Top 5 by Average Unit Price (Profit Margin Proxy)
top_margin = sales_df.groupby('Product')['Unit_Price'].mean().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
sns.barplot(x=top_margin.values, y=top_margin.index, palette="Oranges_d")
plt.title("🏷️ Top 5 Products by Avg Unit Price (Margin Proxy)")
plt.xlabel("Average Unit Price ($)")
plt.tight_layout()
plt.show()
```
